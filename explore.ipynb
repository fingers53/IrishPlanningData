```python
wexford_multi_units = df[(df.NumResidentialUnits > 1) & (df.PlanningAuthority == 'Wexford County Council')]
```

```python
wexford_multi_units.groupby('NumResidentialUnits')['Decision'].value_counts().to_csv('How many plans above X number of houses are approved or refused.csv')
```
